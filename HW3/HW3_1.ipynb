```python
# Description : This code was written by Barış Büyükyılmaz for ELE492 third homework.
import numpy as np
import matplotlib.pyplot as plt

from skimage import img_as_uint
from skimage.io import imshow, imread
from skimage.color import rgb2hsv
from skimage.color import rgb2gray
```

```python
photo = imread('/content/ELE492---Image-Processing-Homeworks/HW3/photo.jpeg')   # Read image
imshow(photo)                                                                   # Display image
```

```python
photo.shape
```

```python
fig, ax = plt.subplots(1, 3, figsize=(6,4), sharey=True)  # Create 3 subplots

slice_with = int(photo.shape[1]/3)                        # Store image with as int

ax[0].imshow(photo[:, 0 : slice_with])                    # Display columns 0 to slice_with 
ax[0].set_title('Left')                                   # Set title as Left

ax[1].imshow(photo[:, slice_with : slice_with * 2])       # Display columns slice_with to 2 times slice_with 
ax[1].set_title('Middle')                                 # Set title as Middle

ax[2].imshow(photo[:, slice_with * 2 : ])                 # Display columns 2 times slice_with to end 
ax[2].set_title('Right')                                  # Set title as Right
```

```python
imshow(photo[int(photo.shape[0]/3) : int(photo.shape[0]*2/5), int(photo.shape[1]/3) : int(photo.shape[1]*2/3)])   # Display image rows height/3 to height*2/5 and colums width/3 to width*2/3
```

```python
fig, ax = plt.subplots(1, 3, figsize=(12,4), sharey=True)     # Create 3 subplots

ax[0].imshow(photo[:,:,0], cmap='Reds')                       # Display Red values of image
ax[0].set_title('Red')                                        # Set title Red

ax[1].imshow(photo[:,:,0], cmap='Greens')                     # Display Green values of image
ax[1].set_title('Green')                                      # Set title Green

ax[2].imshow(photo[:,:,0], cmap='Blues')                      # Display Blue values of image
ax[2].set_title('Blue')                                       # Set title Blue
```

```python
photo_hsv = rgb2hsv(photo)                                  # Convert image RGB to HSV
fig, ax = plt.subplots(1, 3, figsize=(12,4), sharey=True)   # Create 3 subplots

ax[0].imshow(photo_hsv[:,:,0], cmap='hsv')                  # Display Hue values
ax[0].set_title('Hue')                                      # Set title Hue

ax[1].imshow(photo_hsv[:,:,1], cmap='hsv')                  # Display Saturation Values
ax[1].set_title('Saturation')                               # Set title Saturation

ax[2].imshow(photo_hsv[:,:,2], cmap='hsv')                  # Display Value Values
ax[2].set_title('Value')                                    # Set title Value
```

```python
photo_gray = rgb2gray(photo)                                # Convert image rgb to gray

fig, ax = plt.subplots(1, 5, figsize=(17,9), sharey=True)   # Create 3 subplots

ax[0].imshow(photo_gray, cmap='gray')                       # Display gray image
ax[0].set_title('Grayscale Original')                       # Set title

ax[1].imshow(img_as_uint(photo_gray > 0.25), cmap='gray')   # Display image that is pixel value greater than 0.25
ax[1].set_title('Greater than 0.25')                        # Set title

ax[2].imshow(img_as_uint(photo_gray > 0.5), cmap='gray')    # Display image that pixel value greater than 0.50
ax[2].set_title('Greater than 0.50')                        # Set title

ax[3].imshow(img_as_uint(photo_gray > 0.75), cmap='gray')   # Display image that pixel value greater than 0.70
ax[3].set_title('Greater than 0.75')                        # Set title

ax[4].imshow(img_as_uint(photo_gray > np.mean(photo_gray)), cmap='gray')  # Display image that pixel value greater than mean
ax[4].set_title('Greater than Mean')                                      # Set title
```
