```python
# Description : This code was written by Baris Buyukyilmaz for ELE492 third homework.
import numpy as np
import matplotlib.pyplot as plt

from skimage import img_as_uint
from skimage.io import imshow, imread
from skimage.color import rgb2hsv
from skimage.color import rgb2gray
```

```python
photo = plt.imread('/content/ELE492---Image-Processing-Homeworks/HW3/photo.jpeg')

imshow(photo)
```

```python
photo_gray = rgb2gray(photo)

imshow(photo_gray)
```

```python
plt.hist(photo_gray.ravel(), bins=256)
```

```python
fig, axes = plt.subplots(1, 4, figsize=(17,9), sharey=True)

ax = axes.ravel()
ax[0] = plt.subplot(1, 3, 1)
ax[1] = plt.subplot(1, 3, 2)
ax[2] = plt.subplot(1, 3, 3, sharex=ax[0], sharey=ax[0])

t = 0.5
binary_mask = photo_gray < t
inverted_binary_mask = photo_gray > t


ax[0].imshow(photo_gray, cmap=plt.cm.gray)
ax[0].set_title('Original Grayscale')

ax[1].hist(photo_gray.ravel(), bins=256)
ax[1].set_title('Histogram')
ax[1].axvline(t, color='r')

ax[2].imshow(binary_mask, cmap='gray')
ax[2].set_title('Threshold Grayscale')





```
