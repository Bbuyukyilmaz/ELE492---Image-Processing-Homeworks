```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

img = cv.imread('/content/ELE492---Image-Processing-Homeworks/HW5/mouse_adapter.jpeg', cv.IMREAD_GRAYSCALE)  # Read image


th = cv.adaptiveThreshold(img,255,cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY,11,4)     # Apply adaptive threshold using binary threshold 11x11 kernel

titles = ['Original Image','Adaptive Thresholding']       # Set title for images
images = [img, th]    # Create image list

for i in range(2):
    plt.subplot(2,2,i+1),plt.imshow(images[i],'gray')  # Plot images
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```
