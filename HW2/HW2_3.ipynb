```python
# Description : This code was written by Baris Buyukyilmaz for ELE492 second homework.
# The code shows the HSV image of a dog and the RGB image of the same.

from skimage import io
from skimage import color
from skimage import data
from pylab import *

img = io.imread("/content/ELE492---Image-Processing-Homeworks/HW2/dog.jpg")   # Read image

img_hsv = color.rgb2hsv(img)          # Convert RGB to HSV
img_rgb = color.hsv2rgb(img_hsv)      # Convert HSV to RGB

figure(0)       
io.imshow(img_hsv)                    # Show HSV image as figure 0
figure(1)
io.imshow(img_rgb)                    # Show RGB image as figure 1


```
