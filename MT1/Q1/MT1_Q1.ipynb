```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

# read image
img = cv.imread('/content/ELE492---Image-Processing-Homeworks/MT1/q1_image.png') 

# convert to RGB color-space             
rgb_img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

# convert from RGB color-space to YCrCb
ycrcb_img = cv.cvtColor(rgb_img, cv.COLOR_BGR2YCrCb)

# equalize the histogram of the Y channel
ycrcb_img[:, :, 0] = cv.equalizeHist(ycrcb_img[:, :, 0])

# convert back to RGB color-space from YCrCb
equalized_img = cv.cvtColor(ycrcb_img, cv.COLOR_YCrCb2BGR)

# display original image and equalized image
fig, ax = plt.subplots(1, 2)
ax[0].imshow(rgb_img)
ax[0].set_title("Original Image")
ax[1].imshow(equalized_img)
ax[1].set_title("Equalized Image")
plt.show()
```
